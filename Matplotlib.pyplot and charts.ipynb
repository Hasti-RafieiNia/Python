```python
x = df["month_number"]
a = df["bathingsoap"]
b = df["facecream"]
c = df["facewash"]
d = df["moisturizer"]
e = df["shampoo"]
f = df["toothpaste"]

SaleData = [a, b, c, d, e, f]
colors = ['yellow', 'orange', 'red', 'pink', 'purple', 'blue']
ProductName = ['Bathing Soap', 'Face Cream', 'Face Wash', 'Moisturizer', 'Shampoo', 'Tooth Paste']
fig, ax = plt.subplots()
boxes = ax.boxplot(SaleData , patch_artist = True , labels = ProductName)

for box, color in zip(boxes['boxes'],colors):
    box.set(facecolor = color)
    box.set(edgecolor = 'black')

plt.yticks(size = 10)
ax.set_xticklabels(ProductName, size = 10)
plt.ylabel("Sale Units in Number", size = 15)
plt.title("Sales Data", size  = 20)
plt.show()
```

```python
x = df["month_number"]
a = df["bathingsoap"]
b = df["facecream"]
c = df["facewash"]
d = df["moisturizer"]
e = df["shampoo"]
f = df["toothpaste"]

a = (a - a.mean())/ a.std() 
b = (b - b.mean())/ b.std() 
c = (c - c.mean())/ c.std() 
d = (d - d.mean())/ d.std() 
e = (e - e.mean())/ e.std() 
f = (f - f.mean())/ f.std() 

arr = pd.DataFrame({"Bathing Soap" : a,"Face Cream": b , "Face Wash":c ,"Moisturizer":d,"Shampoo":e,"Tooth Paste":f})

SaleData = [a, b, c, d, e, f]
sns.heatmap(arr, annot = True, cmap = 'crest')
plt.xticks(size = 9, rotation = 0)
plt.yticks(size = 10)
plt.title("Sales Data", size = 25)
plt.ylabel("Month", size = 20)
plt.xlabel("Product", size = 20)
plt.show()
```

**Q07**

```python
x = df["month_number"]
a = df["bathingsoap"]
b = df["facecream"]
c = df["facewash"]
d = df["moisturizer"]
e = df["shampoo"]
f = df["toothpaste"]

sa = sum(a)
sb = sum(b)
sc = sum(c)
sd = sum(d)
se = sum(e)
sf = sum(f)


SaleData = [sa, sb, sc, sd, se, sf]
total = sum(SaleData)
percentages = [( i / total) * 100 for i in SaleData ]
ProductName = ['Bathing Soap', 'Face Cream', 'Face Wash', 'Moisturizer', 'Shampoo', 'Tooth Paste']
colors = ['yellow', 'orange', 'red', 'pink', 'purple', 'blue']
explode = [0, 0, 0, 0, 0, 0]
plt.pie(percentages, explode=explode, labels = ProductName, colors=colors, autopct='%1.1f%%', startangle=90)
#circle = plt.Circle((0, 0), 0.6, color='white')
#plt.gca().add_artist(circle)
plt.title("Sales Data", size  = 20)
plt.legend()
plt.show()
```
